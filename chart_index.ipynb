```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
soup = BeautifulSoup(requests.get('https://kworb.net/spotify/').text)
```

```python
link_list = [l['href'] for l in soup.find('table').find_all('a', href=True) 
 if ('weekly' in l['href']) and ('totals' not in l['href'])]
```
